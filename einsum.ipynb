```python
import torch

A = torch.arange(2*3*4*5).reshape(5, 4, 2, 3)
A = A[2]
A
```

```python
torch.einsum('c w h-> h w c',A)
```

```python
L = torch.Tensor(3, 5, 4, 2) # (1 3 2 0)
for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        for k in range(A.shape[2]):
            for l in range(A.shape[3]):
                L[j][l][k][i] = A[i][j][k][l]
L.flatten()
```

```python

A_dim = [2, 3, 4, 5]
A_stride = [60, 20, 5, 1]

L_dim = [3, 5, 4, 2]
L_stride = [ 40, 8, 2, 1]
ch_inx = [1, 3, 2, 0]
tmp_inx = [3, 0, 2, 1]
```

```python
V = A.reshape(-1)
L = torch.Tensor((120))
for inx in range(len(V)):# ten->T 전체의 넓이
    new_inx = 0
    tmp = inx
    for i in range(len(A_dim)): # dim_num
        print((tmp // A_stride[i]),"*",L_stride[ch_inx[i]],end=" ")
        new_inx += (tmp // A_stride[i]) * L_stride[tmp_inx[i]]
        tmp %= A_stride[i]

    print(new_inx)
    L[new_inx] = V[inx]
    
```
